```python
df2 = pd.read_csv('countries.csv')
us = df2[df2.country == 'United States']
ch = df2[df2.country == 'China']
plt.plot(us.year, (us.population / us.population.iloc[0] * 100)-100)
plt.plot(ch.year, (ch.population/ ch.population.iloc[0] * 100)-100)
plt.legend(['United States', 'China'])
plt.title('China vs US Population Growth')
plt.xlabel('Year')
plt.ylabel('Population Growth %')
plt.show()

```
